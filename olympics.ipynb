```python
#import libraries
import pandas as pd
```

```python
#import dataset
athletes_data = pd.read_csv('athletes.csv')
```

```python
athletes_data.head()
```

```python
#inspection of the data types and the number of non-null rows per column
athletes_data.info()
```

```python
#inspect the number of missing values per column
athletes_data.isna().sum()
```

```python
#fill Null value in the columns with "Null"
athletes_data['nationality'] = athletes_data['nationality'].fillna('Null')
athletes_data['nationality_full'] = athletes_data['nationality_full'].fillna('Null')
athletes_data['nationality_code'] = athletes_data['nationality_code'].fillna('Null')
athletes_data['weight'] = athletes_data['weight'].fillna(athletes_data['weight'].mean())
athletes_data['birth_place'] = athletes_data['birth_place'].fillna('Null')
athletes_data['birth_country'] = athletes_data['birth_country'].fillna('Null')
athletes_data['residence_country'] = athletes_data['residence_country'].fillna('Null')
athletes_data['nickname'] = athletes_data['nickname'].fillna('Null')
athletes_data['hobbies'] = athletes_data['hobbies'].fillna('Null')
athletes_data['occupation'] = athletes_data['occupation'].fillna('Null')
athletes_data['education'] = athletes_data['education'].fillna('Null')
athletes_data['family'] = athletes_data['family'].fillna('Null')
athletes_data['lang'] = athletes_data['lang'].fillna('Null')
athletes_data['coach'] = athletes_data['coach'].fillna('Null')
athletes_data['reason'] = athletes_data['reason'].fillna('Null')
athletes_data['hero'] = athletes_data['hero'].fillna('Null')
athletes_data['influence'] = athletes_data['influence'].fillna('Null')
athletes_data['philosophy'] = athletes_data['philosophy'].fillna('Null')
athletes_data['sporting_relatives'] = athletes_data['sporting_relatives'].fillna('Null')
athletes_data['ritual'] = athletes_data['ritual'].fillna('Null')
athletes_data['other_sports'] = athletes_data['other_sports'].fillna('Null')
```

```python
# Save the dataframe to a CSV file
athletes_data.to_csv('clean/athletes.csv', index=False)
```
